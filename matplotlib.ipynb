```python
#giving labels and title
import matplotlib.pyplot as plt 
plt.plot([1,5,3,4,6,8],[1,5,3,4,3,4])
plt.xlabel('numbers in x-axis')
plt.ylabel('numbers in y-axis')
plt.title('Just a title')
plt.show()
```

```python
#Giving legend for two different lines 
import matplotlib.pyplot as plt 
x1=[1,2,3,4,5,6]
y1=[6,5,4,3,2,1]
x2=[9,6,3]
y2=[4,3,6]
plt.plot(x1,y1,label= "first line")
plt.plot(x2,y2, label="second line")
plt.title('just a title')
plt.legend()
plt.show()
```

```python
#example of bargraph
import matplotlib.pyplot as plt 
x1=[1,2,3,4,5,6]
y1=[6,3,4,3,9,1]
plt.bar(x1,y1,label= "Bar1", color='r')
plt.title('just a title')
plt.legend()
plt.show()
```

```python
#histogram, similar to the bargraph
import matplotlib.pyplot as plt 
x1=[1,2,3,4,5,6,12,15,12,3,99,74,45,65,25,68,40,55,33,66,22,95,85,75,66,66,22,45]
bins= [0,20,40,60,80,100]
plt.hist(x1, bins, histtype='bar', rwidth=0.8)
plt.title('just a title')
plt.show()
```

```python
#plotting sactter graph, look at the marker 
import matplotlib.pyplot as plt 
x1=[1,2,3,4,5,6,1,5,6,9,8,7,4,5,3,6,9,4,1,5,2,6,3,2,5,4]
y1=[6,3,4,3,9,1,4,5,8,7,2,1,5,6,0,0,0,1,5,2,4,5,9,8,7,1]
plt.scatter(x1,y1,label= 'just a random data', color='r', marker='*')
plt.title('just a title')
plt.legend()
plt.show()
```

```python
#plotting scatter plot, here i plot my daily activities
import matplotlib.pyplot as plt
days=[1,2,3,4,5]
sleeping=[6,7,7,9,11]
working= [10,8,9,4,2]
eating=[3,3,3,3,4]
playing=[3,2,4,5,8]
plt.plot([],[], color='m', label='sleeping')
plt.plot([],[], color='r', label='working')
plt.plot([],[], color='k', label='eating')
plt.plot([],[], color='g', label='playing')
plt.stackplot(days,sleeping,working,eating,playing, colors=['m','r','k','g'])
plt.legend()
plt.show()
```

```python
#here i show my activities in sunday slice 
import matplotlib.pyplot as plt
days=[1,2,3,4,5]
sleeping=[6,7,7,9,11]
working= [10,8,9,4,2]
eating=[3,3,3,3,4]
playing=[3,2,4,5,8]
slices=[11,2,4,8]
activities=['sleeping','working','eating','playing']
plt.pie(slices,labels=activities,shadow= True, startangle=90, explode=(0,0.4,0.3,0.2), autopct="%1.1f%%")
plt.title('What i do in sunday?')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import style
style.use('fivethirtyeight')
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
def animate(i):
    graph_data = open('example.txt', 'r').read()
    lines = graph_data.split('\n')
    xs = []
    ys = []
    for line in lines:
        if len(line) > 1:
            x,y = line.split(',')
            xs.append(x)
            ys.append(y)
    ax1.clear()
    ax1.plot(xs,ys)
ani = animation.FuncAnimation(fig,animate, interval = 1000)
plt.show()
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

m = Basemap(projection='mill',
            llcrnrlat = 25,
            llcrnrlon = -130,
            urcrnrlat = 50,
            urcrnrlon = -60,
            resolution='l')

m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='b')
```
